```python
import os
import numpy as np

def load_npz_data(filepath):
    with np.load(filepath, allow_pickle=True) as data:
        return dict(data)

def compare_npz_files(file1, file2):
    data1 = load_npz_data(file1)
    data2 = load_npz_data(file2)

    if data1.keys() != data2.keys():
        return False

    for key in data1:
        if not np.array_equal(data1[key], data2[key]):
            return False

    return True

# Set your two directories here
dir1 = 'Data/n=6_all_with_inductive_connectedness_criterion_forward_connected_decomp/'
dir2 = 'Data/n=6_all_with_inductive_connectedness_criterion_forward_connected_test/'

# List common files
files1 = set(f for f in os.listdir(dir1) if f.endswith('.npz'))
files2 = set(f for f in os.listdir(dir2) if f.endswith('.npz'))

common_files = sorted(files1 & files2)
missing_in_1 = sorted(files2 - files1)
missing_in_2 = sorted(files1 - files2)

# Report missing files
if missing_in_1:
    print("Missing in dir1:", missing_in_1)
if missing_in_2:
    print("Missing in dir2:", missing_in_2)

# Compare common files
for fname in common_files:
    path1 = os.path.join(dir1, fname)
    path2 = os.path.join(dir2, fname)
    if compare_npz_files(path1, path2):
        print(f"[✓] {fname} is identical")
    else:
        print(f"[✗] {fname} differs")
```

```python
import json

def load_jsonl_labels(path):
    with open(path, 'r') as f:
        return [int(line.strip()) for line in f if line.strip()]

def load_json_labels(path):
    with open(path, 'r') as f:
        return json.load(f)

def compare_label_lists(path_jsonl, path_json):
    jsonl_labels = load_jsonl_labels(path_jsonl)
    json_labels = load_json_labels(path_json)

    len1 = len(jsonl_labels)
    len2 = len(json_labels)
    min_len = min(len1, len2)

    print(f"✅ Length of labels.jsonl : {len1}")
    print(f"✅ Length of labels.json  : {len2}\n")

    mismatches = []
    for i in range(min_len):
        if jsonl_labels[i] != json_labels[i]:
            mismatches.append((i, jsonl_labels[i], json_labels[i]))

    if mismatches:
        print(f"❌ Found {len(mismatches)} mismatches:")
        for i, a, b in mismatches:
            print(f"  - Index {i}: JSONL = {a}, JSON = {b}")
    else:
        print("✅ All matching up to minimum length.")

    if len1 != len2:
        print(f"\n⚠️ Length mismatch:")
        if len1 > len2:
            print(f"  JSONL has {len1 - len2} extra entries starting from index {len2}")
        else:
            print(f"  JSON has {len2 - len1} extra entries starting from index {len1}")

```
